```python
import json
import os
import time
from elevenlabs.client import ElevenLabs
from pathlib import Path

# 1. הגדרת המפתח והחיבור
API_KEY = "1e49812863201f3ddaf6124e5cc539596e87e526154a818aa670649b5983f66d"
client = ElevenLabs(api_key=API_KEY)

# 2. הגדרת התיקיות
INPUT_DIR = "/content/drive/MyDrive/selected_clips_hebrew"
OUTPUT_DIR = "/content/drive/MyDrive/elevenlabs_fake"

# יצירת התיקייה החדשה ב-Colab 
os.makedirs(OUTPUT_DIR, exist_ok=True)

TARGET_VOICE_ID = "SAz9YHcvj6GT2YYXdXww"   # River
MODEL_ID = "eleven_multilingual_sts_v2"    # המודל שעבד לנו

# 3. מציאת כל קבצי השמע בתיקייה המקורית
audio_files = [f for f in os.listdir(INPUT_DIR) if f.endswith(('.mp3', '.wav', '.m4a', '.ogg'))]
total = len(audio_files)

if total == 0:
    print("לא נמצאו קבצי שמע בתיקייה. ודאי שכתבת את שם התיקייה המקורית נכון ב-INPUT_DIR.")
else:
    print(f"נמצאו {total} קבצים לעיבוד. מתחיל בלולאה...\n")

    # 4. הלולאה שעוברת קובץ קובץ
    for i, filename in enumerate(audio_files, 1):
        input_path = Path(INPUT_DIR) / filename
        output_filename = f"fake_{filename}"
        output_path = Path(OUTPUT_DIR) / output_filename

        print(f"[{i}/{total}] מעבד את: {filename}...")

        try:
            with open(input_path, "rb") as f_in:
                # הזיוף עם ההגדרות שעבדו לנו לעברית
                audio = client.speech_to_speech.convert(
                    voice_id=TARGET_VOICE_ID,
                    audio=f_in,
                    model_id=MODEL_ID,
                    voice_settings=json.dumps({
                        "stability": 0.45,
                        "similarity_boost": 0.8,
                        "style": 0.0,
                        "use_speaker_boost": True
                    })
                )

                # שמירת הקובץ המזויף בתיקיית ה-FAKE
                with open(output_path, "wb") as f_out:
                    if isinstance(audio, (bytes, bytearray)):
                        f_out.write(audio)
                    else:
                        for chunk in audio:
                            if chunk:
                                f_out.write(chunk)

            print(f"   v נשמר בהצלחה כ- {output_filename}")

        except Exception as e:
            print(f"   x שגיאה בקובץ {filename}: {str(e)}")

        # המתנה של שנייה כדי שהשרת של ElevenLabs לא יחסום אותנו על עומס בקשות
        time.sleep(1)

    print("\nהלולאה סיימה! כל הקבצים המזויפים מחכים לך בתיקיית FAKE.")
```
